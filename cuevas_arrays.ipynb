```python
from google.colab import runtime

#input: single element
def check_value_type (val):
  isNum = val.isnumeric()
  if(isNum):
    return "Numeric"
  else:
    return "String"
    
#input: array   
def check_values (val):
  first = val[0].isnumeric()

  for i in val:
    if(first != i.isnumeric()):
      print("ERROR: Values should be of the same type!")
      runtime.unassign()
    
  if(first):
    return "Numeric"
  else:
    return "String"

def num_array_type(arr):
  arr = [eval(i) for i in arr]
  return arr

def print_exercises():
  print("Select an exercise:")
  print("(1)   Reverse the elements of the array")
  print("(2)   Merge two sorted arrays into one")
  print("(3)   Insert a new element into a sorted array")
  print("(4)   Find the minimum in an array and delete it from the array")
  return input().strip()

def reverse_array(val):
  n = len(val) #size of array
  if(n%2 == 1):
    iterations = (n-1)/2    #middle element will not swap
  else:   
    iterations = n/2        #all elements have to swap
    
  #swap elements to reverse the order
  for i in range(int(iterations)):
    x = val[i]
    val[i] = val[n - (1+i)]
    val[n - (1+i)] = x

  print("Reversed Array:")
  print(val)  

def merge_arrays(arr1, arr2):
  arr3 = []
  x = 0   #array1 pointer
  y = 0   #array2 pointer

  while x < len(arr1) and y < len(arr2):
    if(arr1[x] < arr2[y]):
      arr3.append(arr1[x])
      x += 1
    elif(arr1[x] > arr2[y]):
      arr3.append(arr2[y])
      y += 1
    elif(arr1[x] == arr2[y]):
      arr3.append(arr1[x])
      arr3.append(arr2[y])
      x += 1
      y += 1

  if(x == len(arr1)):
    for i in range(y, len(arr2)):
        arr3.append(arr2[i])
  elif(y == len(arr2)):
    for i in range(x, len(arr1)):
        arr3.append(arr1[i])
          
  print("\nMerged Array:")
  print(arr3)


def insert_element(arr, el):
  n = len(arr)    #no. of array elements
  lb = 0          #lower bound index, first element
  up = n - 1      #upper bound index, last element
  ins = False     #if element is inserted, used as a condition for while loop

  #index of middle element
  mid = int(find_middle(lb, up))
  while ins == False:
    if(arr[mid] < el): #right side
      if(mid + 1 == up): #if mid element is next to upper bound
        arr.insert(mid + 1, el) #insert the element next to the middle element
        ins = True
      else:
        lb = mid
        mid = find_middle (lb, up)
    elif(arr[mid] > el): #left side
      if(mid - 1 == lb): #if mid element is next to upper bound
        arr.insert(mid - 1, el) #insert the element next to the middle element
        ins = True
      else:
        up = mid
        mid = find_middle (lb, up)
    else:
        arr.insert(mid, el) #if element is same as middle, insert it at that position
  
  print("\nArray with new element")
  print(arr)

#input: no of elements
def find_middle(lb, up):
  i = up - lb

  if(i%2 == 0):
    step = int(i/2)
  else:
    step = int((i-1)/2)
  mid = lb + step

  return mid

def find_minimun(arr, type):
  min = arr[0]
  
  for val in arr:
      if(type == "Numeric"):
          if(int(val) < int(min)):
              min = val
      else:
          if(val < min):
              min = val   

  arr.remove(min)
  print("\nMinimum value: " + min)
  print("\nMinimum value deleted: ")
  print(arr)

print("===========  Programming Log 1   ===========")
k = print_exercises()

if(k == "1"):
  print("--- Reverse the elements of the array ---")
  
  print("Enter a set of values (seperate by space):")
  arr = input().split()
  arr_type = check_values(arr)
  
  print("\nArray:")
  print(arr)
  
  reverse_array(arr)
elif (k == "2"):
  print("--- Merge two sorted arrays into one ---")
  
  print("Enter a set of values for array 1 (seperate by space):")
  arr1 = input().split()
  arr1_type = check_values(arr1)
  
  print("Enter a set of values for array 2 (seperate by space):")
  arr2 = input().split()
  arr2_type = check_values(arr2)

  if(arr1_type != arr2_type):
    print(" ERROR: Values of both arrays should be of the same type!")
    runtime.unassign()
  
  if(arr1_type == "Numeric"):
    arr1 = num_array_type(arr1)
    arr2 = num_array_type(arr2)
  
  #print both arrays
  arr1.sort()
  print("\nArray 1:")
  print(arr1)
  arr2.sort()
  print("\nArray 2:")
  print(arr2)

  merge_arrays(arr1, arr2)

elif (k == "3"):
  print("--- Insert a new element into a sorted array ---")
  print("Enter a set of values (seperate by space):")
  arr = input().split()
  arr_type = check_values(arr)

  print("Enter a new element:")
  newEl = input().strip()
  newEl_type = check_value_type(newEl)

  if(arr_type != newEl_type):
    print("ERROR: New element should be of the same type as the elements of the array (" + arr_type + ")!")
    runtime.unassign()

  if(arr_type == "Numeric"):
    arr = num_array_type(arr)
    newEl = int(newEl)

  arr.sort()
  print("\nArray:")
  print(arr)
  print("New Element:")
  print(newEl)
  insert_element(arr, newEl)

  print("-----------------------------") 
elif (k == "4"):
  print("--- Find the minimum in an array and delete it from the array ---")
  print("Enter a set of values (seperate by space):")
  arr = input().split()
  arr_type = check_values(arr)         
  
  print("\nArray:")
  print(arr)

  find_minimun(arr, arr_type)

  print("---------------------------------") 
else:
  print("Command not recognized.")

print("************************")


```
