**SUGGESTED FRIENDS FOR A PERSON IN A SOCIAL NETWORK**

```python
import matplotlib.pyplot as plt
import networkx as nx

socials = nx.Graph()                                                                       # create a social network graph

socials.add_nodes_from(['John',                                                            # add nodes to graph i.e. the people in the social network
                        'Mary', 
                        'Joseph', 
                        'Elizabeth', 
                        'Mark', 
                        'Alexis', 
                        'Marie', 
                        'Amy', 
                        'Daniel', 
                        'Bianca', 
                        'Matthew', 
                        'Jacob', 
                        'Sofia', 
                        'Sabrina', 
                        'Jude',
                        'Vivian'
                        ])

pos={                                                                                      # set coordinates for node positions in visualization
    "John":(1,5),
    "Mary":(3.5,8.5),
    "Joseph":(1,1.4),
    "Elizabeth":(5.8,-2),
    "Mark":(7.9,3.6),
    "Alexis":(10.9,5.6),
    "Marie":(3.2,-1),
    "Amy":(8.9,7),
    "Daniel":(10,2.7),
    "Bianca":(6,2.1),
    "Matthew":(3.3,3.4),
    "Jacob":(5.9,6.6),
    "Sofia":(12.2,0),
    "Sabrina":(9,-2),
    "Jude":(7,9),
    "Vivian":(12.5,9)
}

socials.add_edges_from([                                                                   # create edges to connect nodes -- direct friendships
    ('John', 'Joseph'),
    ('John', 'Matthew'),
    ('Joseph', 'Matthew'),
    ('Marie', 'Bianca'),
    ('Mary', 'Matthew'),
    ('Mary', 'Jacob'),
    ('Mary', 'Jude'),
    ('Matthew', 'Bianca'),
    ('Elizabeth', 'Bianca'),
    ('Elizabeth', 'Sabrina'),
    ('Jacob', 'Jude'),
    ('Jacob', 'Bianca'),
    ('Jacob', 'Mark'),
    ('Mark', 'Bianca'),
    ('Mark', 'Amy'),
    ('Mark', 'Sabrina'),
    ('Mark', 'Daniel'),
    ('Alexis', 'Amy'),
    ('Alexis', 'Daniel'),
    ('Alexis', 'Sofia'),
    ('Alexis', 'Vivian')
])

nx.draw(socials,pos=pos,with_labels=True,                                                   # plot social network for visualization
        node_color="green",node_size=2700,
        font_color="white",font_size=10, font_weight="bold",
        width=3)
plt.margins(0.2)
plt.show()

def get_suggested_friends(G, node):
  '''
  Takes in the social network graph and a person(node),
  returns suggested friends list for person(node)
  '''
  mutuals = [n for n in socials.neighbors(node)]                                           # get "mutuals of person" by getting direct neighbors of node
  friends_of_mutuals = []                                                                  # create empty list for "friends_of_mutuals" i.e. all 2nd degree neighbors of a node
  for mutual in mutuals:
    friends_of_one_mutual = [n for n in socials.neighbors(mutual) if n not in mutuals and n is not node]     # add person to list if it is not part of node's friends list
    friends_of_mutuals.append(friends_of_one_mutual)                                       # add list to friends_of_mutuals
  
  # print(f"These are {node}'s friends {mutuals}")
  # print(f"These are each mutual's friends {friends_of_mutuals}")

  suggested_friends_list = ', '.join({person for sublist in friends_of_mutuals for person in sublist})  # create a set of all friends of friends to remove duplicates
  if suggested_friends_list == '':
    print(f"There are no suggested friends for {node}")
  else:
    print(f'The suggested friends for {node} will be', end=' ')
  return suggested_friends_list                                                            # return list of suggested friends for a certain person(node)
```
